```python
import pandas as pd
from pandas import DataFrame as df
```

```python
raw_data = pd.read_csv('Data/cardio_train.csv',sep=';')
raw_data.head()
```

```python
raw_data.shape
```

```python
raw_data.describe()
```

```python
# setup variabel preditor dan variabel response
column_X = ['age','gender','height','weight','ap_hi','ap_lo','cholesterol','gluc','smoke','alco','active']
column_y = ['cardio']

X = raw_data[column_X]
y = raw_data[column_y]
```

```python
df(X).head()
```

```python
df(y).head()
```

```python
# Memecah data train dan test
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(X,y, test_size = 0.20, random_state = 42)

print(train_X.shape)
print(train_y.shape)
```

```python
# Lakukan modeling
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix as cm
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
model = LogisticRegression()
model.fit(train_X, train_y['cardio'])
```

## Prediksi model Logistic Regression

```python
prediction_value = model.predict(test_X)
df(prediction_value).head()
```

```python
test_y['cardio']
```

```python
# Confusion matrix untuk mengetahui berapa True Positive, True Negatice, False Positive dan False Negatice
confusion_matrix = cm(test_y['cardio'], prediction_value)
confusion_matrix
```

```python
print(f'Tingkat keberhasilan dari True Positive dan False Negative {((confusion_matrix[0][0]+confusion_matrix[1][1])/confusion_matrix.sum())*100}')
```

```python
from sklearn.metrics import classification_report
target_names = ['Beresiko', 'Tidak Beresiko']
print(classification_report(test_y, prediction_value, target_names=target_names))
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(confusion_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

## Receiving Operation Characteristic

```python
y_pred_proba = model.predict_proba(test_X)[::,1]
# print(y_pred_proba)
fpr, tpr, threshold = metrics.roc_curve(test_y,  y_pred_proba)
auc = metrics.roc_auc_score(test_y, y_pred_proba)

plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

Area Under Curve sebesar 76%, bisa dikatakan model ini sebagai klasifikasi yang cukup baik
